```python
#imports
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
from pprint import pprint
import json
#2005-2021

MinWage = pd.read_csv("InputData/MinWage.csv")
Milk = pd.read_csv("InputData/MilkPrice.csv")
Egg = pd.read_csv("InputData/Eggprice.csv")
beef = pd.read_csv("InputData/beef.csv")
cpi = pd.read_csv("InputData/annual_cpi.csv")
Bread = pd.read_csv("InputData/Bread.csv")
```

```python
Breadyy = Bread.loc[(Bread["Year"]<=2021)]


Breadyy["mean"] = Breadyy.iloc[:, 1:12].mean(axis = 1)
Breadyy

BreadFrame = pd.DataFrame({
    'Year': Breadyy['Year'],
    'Ave Bread Price': Breadyy['mean']
})
```

```python
#Create a Beef Data Frame
beefDF = pd.DataFrame({
    'Year': beef['Year'],
    'Ave Beef Price (per pound)': beef['Retail value (cents per pound)']
})

#Group the Beef Data by Year
beefDF = beefDF.groupby(beef['Year']).mean()
beefDF['Ave Beef Price (per pound)'] = beefDF['Ave Beef Price (per pound)']/100
beefDF

shortBeefDF = beefDF.loc[(beefDF['Year']>=2005)]

#Remove the year column
del shortBeefDF['Year']
```

```python
#Set the index for the CPI data to “Year”
cpiPlot = cpi.copy()
cpiPlot = cpiPlot.set_index('Year')

#Graph the CPI Data
cpiPlot
```

```python
MilkYear = Milk.loc[(Milk['Year'] >= 2005) & (Milk['Year'] <= 2021)]

MilkYear['mean'] = MilkYear.iloc[:, 1:12].mean(axis=1)

MilkFrame = pd.DataFrame({
    'Year': MilkYear['Year'],
    'Ave Milk Price': MilkYear['mean']
})
```

```python
EggYear = Egg.loc[(Egg['Year'] >= 2005) & (Egg['Year'] <= 2021)]

EggYear['mean'] = EggYear.iloc[:, 1:12].mean(axis=1)

EggFrame = pd.DataFrame({
    'Year': EggYear['Year'],
    'Ave Egg Price(per dozen)': EggYear['mean']
})
```

```python
MinWageYear = MinWage.loc[(MinWage['Year'] >= 2005) & (MinWage['Year'] <= 2021)]
Frame = pd.DataFrame({
    'Year': MinWageYear['Year'],
    'State': MinWageYear['State'],
    'Minimum Wage': MinWageYear['Effective.Minimum.Wage'],
    
})
```

```python
AFrame = Min.merge(MilkFrame, on="Year")
BFrame = AFrame.merge(EggFrame, on="Year")
CFrame = BFrame.merge(shortBeefDF, on="Year")
DoneFrame = CFrame.merge(BreadFrame, on="Year")
XFrame = DoneFrame.set_index("Year")
XFrame
```

```python
FFrame = DoneFrame.merge(cpiPlot, on="Year")
```

```python
#Minimum Wage
FFrame.plot.scatter(x="Annual CPI", y="Average Minimum Wage", c="b", title="Minimum Wage V CPI")
(slope, yIntercept, rvalue, pvalue, sterr) = linregress(FFrame['Annual CPI'], FFrame['Average Minimum Wage'])
#regress_values
values = (FFrame['Annual CPI'] * slope) + yIntercept
plt.plot(FFrame['Annual CPI'], values, 'r')
line = f'y = {slope:.2f}x + {yIntercept:.2f}'
R = f'Correlation Value = {rvalue:.2f}'
plt.annotate(R,(200, 9), fontsize=12, color='r')
plt.annotate(line,(240, 6), fontsize=12, color='r')
print(f"Minimum Wage V CPI correlation: {rvalue}")
plt.savefig('OutputData/Wagecorrelation.png')

#Eggs
FFrame.plot.scatter(x="Annual CPI", y="Ave Egg Price(per dozen)", c="b", title="Egg Price V CPI")
(slope, yIntercept, rvalue, pvalue, sterr) = linregress(FFrame['Annual CPI'], FFrame['Ave Egg Price(per dozen)'])
#regress_values
values = (FFrame['Annual CPI'] * slope) + yIntercept
plt.plot(FFrame['Annual CPI'], values, 'r')
line = f'y = {slope:.4f}x + {yIntercept:.2f}'
R = f'Correlation Value = {rvalue:.2f}'
plt.annotate(R,(200, 2.2), fontsize=12, color='r')
plt.annotate(line,(245, 2), fontsize=12, color='r')
print(f"Egg Price V CPI correlation: {rvalue}")
plt.savefig('OutputData/Eggcorrelation.png')

#Beef
FFrame.plot.scatter(x="Annual CPI", y="Ave Beef Price (per pound)", c="b", title="Beef Price V CPI")
(slope, yIntercept, rvalue, pvalue, sterr) = linregress(FFrame['Annual CPI'], FFrame['Ave Beef Price (per pound)'])
#regress_values
values = (FFrame['Annual CPI'] * slope) + yIntercept
plt.plot(FFrame['Annual CPI'], values, 'r')
line = f'y = {slope:.2f}x + {yIntercept:.2f}'
R = f'Correlation Value = {rvalue:.2f}'
plt.annotate(R,(200, 7), fontsize=12, color='r')
plt.annotate(line,(240, 4.2), fontsize=12, color='r')
print(f"Beef Price V CPI correlation: {rvalue}")
plt.savefig('OutputData/Beefcorrelation.png')

#Bread
FFrame.plot.scatter(x="Annual CPI", y="Ave Bread Price", c="b", title="Bread Price V CPI")
(slope, yIntercept, rvalue, pvalue, sterr) = linregress(FFrame['Annual CPI'], FFrame['Ave Bread Price'])
values = (FFrame['Annual CPI'] * slope) + yIntercept
plt.plot(FFrame['Annual CPI'], values, 'r')
line = f'y = {slope:.2f}x + {yIntercept:.2f}'
R = f'Correlation Value = {rvalue:.2f}'
plt.annotate(R,(200, 2.2), fontsize=12, color='r')
plt.annotate(line,(240, 1.5), fontsize=12, color='r')
print(f"Bread Price V CPI correlation: {rvalue}")
plt.savefig('OutputData/Breadcorrelation.png')

#Milk
FFrame.plot.scatter(x="Annual CPI", y="Ave Milk Price", c="b", title="Milk Price V CPI")
(slope, yIntercept, rvalue, pvalue, sterr) = linregress(FFrame['Annual CPI'], FFrame['Ave Milk Price'])
values = (FFrame['Annual CPI'] * slope) + yIntercept
plt.plot(FFrame['Annual CPI'], values, 'r')
line = f'y = {slope:.4f}x + {yIntercept:.2f}'
R = f'Correlation Value = {rvalue:.2f}'
plt.annotate(R,(200, 3), fontsize=12, color='r')
plt.annotate(line,(240, 3.7), fontsize=12, color='r')
print(f"Milk Price V CPI correlation: {rvalue}")
plt.savefig('OutputData/Milkcorrelation.png')
```
